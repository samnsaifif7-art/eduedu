# EduMind AI — EDA: Student Performance Dataset

Exploratory Data Analysis on the synthetic student records generated by `data/generate_dataset.py`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid', palette='muted')
plt.rcParams['figure.figsize'] = (10, 5)

df = pd.read_csv('../data/sample_students.csv')
print(df.shape)
df.head()
```

```python
df.describe()
```

## Grade Distribution

```python
order = ['F','D','C','B-','B','A-','A+']
df['grade'].value_counts()[order].plot(kind='bar', color=sns.color_palette('muted', 7))
plt.title('Grade Distribution')
plt.xlabel('Grade'); plt.ylabel('Count')
plt.tight_layout(); plt.show()
```

## Correlation Heatmap

```python
numeric = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,7))
sns.heatmap(numeric.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout(); plt.show()
```

## Study Hours vs Final Score

```python
sns.scatterplot(data=df, x='study_hours_per_day', y='final_score', hue='grade',
                hue_order=order, alpha=0.6, palette='Spectral')
plt.title('Study Hours vs Final Score')
plt.tight_layout(); plt.show()
```

## Attendance vs Final Score by Subject

```python
g = sns.FacetGrid(df, col='subject', col_wrap=3, height=3.5)
g.map_dataframe(sns.scatterplot, x='attendance_pct', y='final_score', alpha=0.4)
g.set_titles('{col_name}')
g.tight_layout(); plt.show()
```

## Feature Engineering Preview

```python
df['study_x_attendance'] = df['study_hours_per_day'] * df['attendance_pct'] / 100
df['consistency_score']  = (df['attendance_pct'] + df['assignments_completed_pct']) / 2

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(df['study_x_attendance'], df['final_score'], alpha=0.4, c='steelblue')
axes[0].set_xlabel('study × attendance'); axes[0].set_ylabel('Final Score')
axes[0].set_title('Study×Attendance vs Score')

axes[1].scatter(df['consistency_score'], df['final_score'], alpha=0.4, c='coral')
axes[1].set_xlabel('Consistency Score'); axes[1].set_ylabel('Final Score')
axes[1].set_title('Consistency Score vs Score')

plt.tight_layout(); plt.show()
```
